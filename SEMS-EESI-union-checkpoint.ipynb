```python
#importing modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime as dt
```

```python
#importing EESI and SEMS data
#input date as a string of YYYYMMDD
sampling_date = "20230711"
#input species as a string of its full name, only first letter capitalized
species = "Oxybenzone"

sems_df = pd.read_excel(sampling_date+"_SEMS.xlsx")
"""[Date]_SEMS.xlsx should have two columns: on the left should be 'Time' 
and on the right should be 'Mass' (units of micro- or nanograms per cubic meter)"""

eesi_df = pd.read_excel(sampling_date+"_EESI.xlsx",species)
"""[Date]_EESI.xlsx should have at least one sheet that bears the name of the compound whose
data is on the sheet. The sheet itself should contain two (or three) columns: on the left should be 
'EESI Time', next should be 'Intensity' (after background subtraction and normalization), and finally (and optionally) 
I found a column titled 'Ignore' helpful for converting UTC data to UTC-7 data, so it'd line up right with the 
SEMS data and I wouldn't have to use timedelta."""
```

```python
#finding missing data
e_index = ~np.isnan(eesi_df.iloc[:,1]) #all eesi points that are not NaN
```

```python
#correcting data (i.e., removing missing data)
eesi_datestamp = eesi_df.iloc[:,0][e_index] #eesi time&date (we only want time so we'll change later)
eesi_intensities = eesi_df.Intensity[e_index] #eesi normalized ion intensities 

#converting eesi_datestamp to only timestamps

eesi_time = [] #initialize an empty list 
for n in range(len(eesi_datestamp)):
    """creates a string of the timestamp (HH:MM:SS.S000 format) and splits by colons"""
    try: 
        nlist = str(eesi_datestamp[n].time()).split(':') 
        eesi_time.append(nlist) 
    except KeyError: #KeyError will be raised whenever value at the key is NaN (I'm pretty sure)
        continue #skip iteration if KeyError is raised

eesi_int_data = [] #basically the same process for a list of ion intensity data but without the string splitting stuff
for n in range(len(eesi_intensities)):
    try: 
        eesi_int_data.append(eesi_intensities[n]) #data appended as floats
    except KeyError: #same reason for try/except as above
        continue
        
"""print(np.shape(eesi_time), np.shape(eesi_int_data)) 
-> (x,3) (x,) (can't make an matrix with lists of incompatible shape like that)"""

#we're going to convert the time stamps into seconds since midnight and then convert back down the line
eesi_seconds = []
for i in range(len(eesi_time)):
    timelist = eesi_time[i] #singles out time list in question
    sec_sum = 3600*int(timelist[0]) + 60*int(timelist[1]) + float(timelist[2])  #3600*hours + 60*minutes + seconds
    eesi_seconds.append(sec_sum) #add sum of seconds since midnight to new list

#matrix with seconds since midnight in the 0th (left) column and ion intensity in the 1st (right)
eesi_data = np.matrix([eesi_seconds, eesi_int_data])
```

Now we need to make a dictionary.
Dict keys will be tuples containing 1 SEMS start point and 1 SEMS end point. Except for the first and last points, each point will be used once as an endpoint and once as a start point. Dict values will be lists of ion intensities from EESI points that fall between the start and endpoints. We will need to convert the SEMS times to seconds since midnight first. 

```python
#converting to seconds since midnight
sems_seconds = []  #initialize empty list
for i in range(len(sems_df.iloc[:,0])):
    timelist = str(sems_df.iloc[i,0]).split(':')
    sec_sum = 3600*int(timelist[0]) + 60*int(timelist[1]) + float(timelist[2])  #3600*hours + 60*minutes + seconds
    sems_seconds.append(sec_sum) #add sum of seconds since midnight to new list
mass_array = np.array(sems_df.iloc[:,1])
sems_matrix = np.matrix([sems_seconds, mass_array])
```

Now we have a matrix for the SEMS data (seconds since midnight on the left and mass on the right) and a matrix for the EESI data (seconds since midnight on the left and ion intensity on the right). Let's get to making that dictionary.

```python
#initialize a list, then transpose the matrix to be able to iterate thru it cleanly, change each entry (lists I think?) to an 
#int and put them in a tup together
tuple_list = []
for i in range(len(sems_matrix[0,:].T)-1): 
    """length-1 because we don't want the last point to try to make a tuple with the next one (which doesn't exist)"""
    timerange_tup = int(sems_matrix[0,:].T[i]), int(sems_matrix[0,:].T[i+1])
    tuple_list.append(timerange_tup)

#now we have a tup list which we can iterate thru in order to find which EESI points belong where
```

```python
dict1 = {} #intialize empty dict
for start, end in tuple_list: #unpack tuples in tuple_list, then iterate thru them
    dict1[(start,end)] = []
    for n in range(len(eesi_data[0,:].T)): 
        """iterate over eesi_data timepoints (I know this is in O(n^2) or O(n*m) and is thus not ideal but whatever)"""
        if int(eesi_data[0,n]) > start and int(eesi_data[0,n]) < end:
            dict1[(start,end)].append(eesi_data[1,n])
        else:
            continue
```

Now we have that dictionary. Let's iterate thru it and make a new dictionary with the same keys where the values are the average and standard error of the ion intensities.

```python
#we'll need one scipy function for the standard error
from scipy.stats import sem
dict2 = {}
for n in range(len(dict1.keys())):
    dict1key = list(dict1.keys())[n] #acquiring dictionary key (one tuple) in question
    if len(dict1[dict1key]) != 0:
        """assigning the value of that key in the new dict to be a list
         with mean and standard error of the values in the first dict. 
         Also, python gave me a bunch of weird runtime errors when I wrote this with 
         no if/else (I just wrote what's under the if statement as the second 
         line under the for), and even though it still worked with all the 
         runtime errors I figured I'd just do the if/else to make the functionally meaningless error go away."""
        dict2[dict1key] = [np.mean(dict1[dict1key]), sem(dict1[dict1key])] 
    else:
        dict2[dict1key]= (np.nan, np.nan)
```

```python
#now we'll assign the EESI avg. and SE to the first timestamp in each dict key
eesi_dict = {}
""" so many different ways to do this with arrays or whatever but
I just chose a dict b/c it was what I was most familiar with"""
for time1, time2 in list(dict2.keys()): 
    h = time1//3600                    #converting back to UTC-7
    m = (time1%3600)//60
    s = (time1%3600)%60
    timestamp1 = dt.time(hour=h, minute = m, second = s)
    eesi_dict[timestamp1] = dict2[(time1,time2)] 
    
eesi_avgs = []
eesi_SEs = []
eesi_SEs_nonan = []
sems_times = list(eesi_dict.keys()) #gleaning list of timestamps from eesi_dict keys
"""converting entire list of timestamps to date+time 
(for some reason you can't just graph time with a time object, so today's date is used arbitrarily)"""
datetimes = [dt.datetime.combine(dt.date.today(), t) for t in sems_times] 

for l in eesi_dict.values():
    eesi_avgs.append(l[0])
    eesi_SEs.append(l[1])
    """we need to sort all of the actual data separate from the NaNs b/c NaN 
    is considered greater than the data, and will thus mess with the percentile calculation"""
    if ~np.isnan(l[1]): 
        eesi_SEs_nonan.append(l[1])
    

SE_90th_pct = np.percentile(eesi_SEs_nonan,90)


for n in range(len(eesi_SEs)):
    if  eesi_avgs[n] <= 0: #if the data is negative or zero (should have been background subtracted) just make it NaN
        eesi_SEs[n] = np.nan
        eesi_avgs[n] = np.nan

ax = plt.axes()
ax.plot(datetimes, eesi_avgs)
fmt = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(fmt) #telling python to only show time on x axis, not date
plt.xlabel('Time')
plt.ylabel('Ion intensity')
plt.title(species+" EESI reading vs. time")
plt.show()
```

Our next goal is a graph with ion intensity on the y axis and mass concentration on the x axis (with a fit line). To get there, we'll create some time series arrays and slice them up. Our final goal will be to write to an excel file the following data: time series of SEMS data, averaged EESI data, and EESI standard errors.

```python
"""because we assigned the average EESI ion intensity
to the first time in the window, we nixed 1 data point (last mass concentration point doesn't get
an associated avg. ion intensity), so we'll tack a NaN onto the end. Helpfully, we always end experiments on a blank,
so the last values will be NaN anyway, so we don't have to worry about diluting the data or whatever."""
eesi_avgs.append(np.nan)
```

```python
eesi_SEs.append(np.nan)
```

```python
sems_df['Intensity'] = eesi_avgs #concatenating avg. EESI data to the SEMS dataframe with time and mass concentration
""" #for some reason I kept getting inconsistent errors with regards to
the length of eesi_SEs it wanted so I just did this quick try/except to appease it"""
try:
    sems_df['SE'] = eesi_SEs
except ValueError:
    sems_df['SE'] = eesi_SEs[:-1]
    
x = datetimes #datetime points from earlier (because you can't graph time with just a timestamp, who knows why)
y1 = sems_df.Mass[:-1] #omitting last point to make it fit better
y2 = sems_df.Intensity[:-1]

fig, ax1 = plt.subplots() #creating plot

fmt = mdates.DateFormatter('%H:%M') #telling it to actually graph with just time, not date too
ax1.xaxis.set_major_formatter(fmt)

ax2 = ax1.twinx() #creating secondary axis
ax1.plot(x, y1, 'g-') #assigning mass concentration vs. time to the primary y axis and coloring it green
ax2.plot(x, y2, 'b-') #assigning ion intensity vs. time to the secondary y axis and coloring it blue

ax1.set_xlabel('Time') #labeling axes and showing plot
ax1.set_ylabel('Mass concentration (ug m^-3)', color='g')
ax2.set_ylabel('Ion intensity (ions/sec)', color='b')
plt.title(species+' EESI and SEMS data vs. time')
plt.show()
```

```python
sems_column = np.asarray(sems_df.Mass)
for i in range(len(sems_column)): 
    """#removing all the mass data above a certain mass loading 
    (we don't expect natural concentrations of 800 ug m^-3 or whatever, so we omit them from the cal. curve)"""
    if sems_column[i] >= 200:
        sems_column[i]=np.nan
```

```python
x = pd.DataFrame(sems_column, columns = ['Mass']).Mass 
y = sems_df.Intensity #isolating column of ion intensity data from new SEMS dataframe
"""#I won't use p and t until making the dataframe to write 
to the excel, but they fit the pattern with x and y so I put them here"""
p = sems_df.SE 
t = sems_df.Time
indexes = ~np.isnan(y) & ~np.isnan(x) #once again finding values where there is ion intensity data (not background)

#now, time to do the regression
z = np.array([x[indexes],y[indexes]],dtype=float) #make a matrix with both series of data
cov_mat = np.cov(z) #covariance matrix pertaining to the matrix z

slope = cov_mat[0,1]/cov_mat[0,0] #gleaning slope of regression line from covariance matrix
intercept = np.mean(y)-slope*np.mean(x)
"""I just did this so it would show up like a normal line on the graph, 
other ways are possible but this came to mind first"""
a = range(int(max(x))) 
b = slope*a + intercept

#time to make the correlation matrix
big_mat = pd.concat([x[indexes],y[indexes]], axis=1) 
big_mat.columns = ['Mass','Intensity'] 
c = big_mat.corr()
r = c.iloc[0,1] #gleaning correlation coefficient R from correlation matrix 

plt.plot(a,b, 'red', linestyle='dashed')

try: #I had the same inconsistent error problem as earlier, so it's the same simple try/except to address it
    plt.errorbar(sems_df.Mass, sems_df.Intensity, yerr=np.asarray(eesi_SEs), fmt='o', capsize = 2.5)
except ValueError:
    plt.errorbar(sems_df.Mass, sems_df.Intensity, yerr=np.asarray(eesi_SEs)[:-1], fmt='o', capsize = 2.5)

plt.scatter(x,y)
plt.xlabel('Mass concentration (ug m^-3)')
plt.ylabel('Ion intensity (ions/sec)')
plt.title(species+' EESI data vs. SEMS data')
plt.show()
print("R^2 is {}, R is {}. Regression line equation is y = {}*x + {}.".format(round(r**2, 3), round(r, 3),round(slope,8), round(intercept,6)))
```
